#### Testing importing GZIP files

```python
import gzip
import json
```

#### Four important transformation steps (in backwards order)
1. Python Data Structures
2. Python Strings containing serialized representation (JSON)
3. List of Bytes containing a representation of the string ("UTF-8")
4. List of Bytes containing a representation of the previous byte list ("gzip")

```python
def tweet_gen(jsonfilename):
    with gzip.GzipFile(jsonfilename, 'r') as fin:
        for line in fin:
            json_str = line.decode('utf-8')
            yield json.loads(json_str)
def read_gzip(jsonfilename):
    data = []
    with gzip.GzipFile(jsonfilename, 'r') as fin:
        for line in fin:
            json_str = line.decode('utf-8')
            data.append(json.loads(json_str))
    return data
```
